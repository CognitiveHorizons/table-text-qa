```python
from transformers import BertTokenizerFast
tokenizer = BertTokenizerFast.from_pretrained("/mnt/infonas/data/yashgupta/save/bert_base_logits_top/2021_08_13_11_42_07/checkpoint-epoch0")
```

```python
from tqdm.notebook import tqdm
cnt_old = 0
cnt_new = 0
cnt = 0
for y, x in enumerate(train_bf):
    ans = x['answer-text']
    row_str = ""
    for c,r in x['table_row'].items():
        row_str+=str(c)+" is "+str(r)+" . "
    old = tokenizer.convert_tokens_to_string(tokenizer.tokenize(row_str + x["table_passage_row_old"])[:510])
    new = tokenizer.convert_tokens_to_string(tokenizer.tokenize(row_str + x["table_passage_row"])[:510])
    b1 = False, b2 = False
    if ans.lower() in old.lower():
        cnt_old += 1
        # old_avg += 
        b1 = True
    if ans.lower() in new.lower():
        cnt_new += 1
        b2 = True
    cnt += 1
    if cnt%1000==0:
        print(cnt, cnt_old, cnt_new)
print(cnt_old, cnt_new)
```

```python
from tqdm.notebook import tqdm
chr_old = 0
cnt_old = 0
chr_new = 0
cnt_new = 0
cnt = 0
for x in tqdm(train_af):
    ans = x['answer-text']
    old = x["table_passage_row_old"]
    new = x["table_passage_row"]
    if ans.lower() in old.lower():
        chr_old += old.lower().find(ans.lower())
        cnt_old += 1
    if ans.lower() in new.lower():
        chr_new += new.lower().find(ans.lower())
        cnt_new += 1
    cnt += 1
    if cnt%1000==0:
        print(cnt, chr_old, chr_new)
print(cnt_old, chr_old, cnt_new, chr_new)
```
